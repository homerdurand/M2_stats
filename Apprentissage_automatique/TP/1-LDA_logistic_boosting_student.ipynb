>Based on the Cholesky decomposition of a $2 \times 2$ covariance matrix $\Sigma$, write a function that generates a multivariate Gaussian $n$-sample of mean $\mu \in \mathbb R^2$ and covariance $\Sigma$.
The corresponding numpy array should be of size $(n, 2)$.

>Compute the mean and the empirical covariance of the sample using Numpy routines.

```python
def gaussianMultivar(n, mu, sigma, theta=0.0) :
    d = len(mu)
    Sigma = covariance(sigma[0], sigma[1], theta)
    samples = []
    for i in range(n) : 
        Z = np.random.normal(0, 1, d)
        L =  np.linalg.cholesky(Sigma)
        samples.append(mu + L.dot(Z))
    return np.array(samples)

def empirical_mean(samples) :
    return np.sum(samples, axis=0)/samples.shape[0]

def empirical_covariance(samples) : 
    n = samples.shape[0]
    mu = empirical_mean(samples)
    return (samples - mu).T.dot((samples - mu))/n

def MSE(theta, theta_hat) :
    return np.mean((theta-theta_hat)**2)
```

>Based on the following code, implement a linear discriminant classifier, taking as parameters an $n \times 2$ Numpy array as data and a size-$n$ array of labels.

```python
# Answer
from sklearn.base import BaseEstimator
from sklearn.discriminant_analysis import LinearClassifierMixin

class LDA(BaseEstimator, LinearClassifierMixin):
    """
        LDA classifier for two classes.
    """
    def __init__(self, prior=None):
        """
            prior: wether to use prior in the intercept. Default is false.
        """
        self.prior = prior
        
    def fit(self, X, y):
        # Estimate covariance matrix and means
        # Todo
        sigma = np.cov(X.T)
        mu1, mu2 = np.mean(X[y==1], axis=0), np.mean(X[y==-1], axis=0)
        sigma_inv = np.linalg.inv(sigma)
        # End todo
        if not self.prior:
            pi1, pi2 = 0.5, 0.5
        else:
            pi1, pi2 = np.mean(y == y.max()), np.mean(y == y.min())
        # Compute direction and intercept
        # Todo
        self.h = (mu1 - mu2).T.dot(sigma_inv)
        self.b = 0.5*(mu2.T.dot(sigma_inv.dot(mu2)) - mu1.T.dot(sigma_inv.dot(mu1))) + np.log(pi1/pi2)
        # End todo
        return self

    def decision_function(self, X):
        # Compute decisions
        # Todo
        decisions = (self.h.dot(X.T) + self.b)
        # End todo
        return decisions

    def predict(self, X):
        # Compute predictions
        # Todo
        decisions = self.decision_function(X)
        predictions = (decisions>=0)*1 + (decisions<0)*(-1)
        # End todo
        return predictions
```

## Fisher discriminant analysis <a id="part1sec3"></a>
>Implement the Fisher discriminant analysis based on the following code.
In practice, what is the difference between LDA and FisherDA?

```python
class FisherDA(BaseEstimator, LinearClassifierMixin):
    """
        Fisher discriminant analysis for two classes.
    """
    def fit(self, X, y):
        pass

    def decision_function(self, X):
        pass

    def predict(self, X):
        pass
```

```python
# Answer
class FisherDA(BaseEstimator, LinearClassifierMixin):
    """
        Fisher discriminant analysis for two classes.
    """
    def fit(self, X, y, showError=True):
        # Estimate prior, covariance matrix and means
        # To do
        pi1, pi2 = np.mean(y == y.max()), np.mean(y == y.min())
        sigma1, sigma2 = np.cov(X[y==y.max(), :].T), np.cov(X[y==y.min(), :].T)  
        sigma = pi1*sigma1 + pi2*sigma2
        sigma_inv = np.linalg.inv(sigma)  # Inverse covariance matrix
        mu1, mu2 = X[y==1, :].mean(axis=0), X[y==-1, :].mean(axis=0)
        # End todo
        
        # Compute direction and intercept
        # Todo
        self.h = sigma_inv.dot(mu1-mu2)
        self.b = 0
        
        ypred = self.decision_function(X)
        ind = np.argsort(ypred)
        err = np.cumsum(y[ind]) + np.sum(y==-1)
        if showError :
            plt.figure()
            plt.plot(ypred[ind], err)  # Error
        iintercept = np.argmin(err)
        if iintercept < y.size-1:
            self.b = -0.5*(ypred[ind[iintercept]] + ypred[ind[iintercept+1]])
        else:
            self.b = -ypred[ind[iintercept]]
        return self

    def decision_function(self, X):
        # Compute decisions
        # Todo
        decisions = X.dot(self.h) + self.b
        # End todo
        return decisions

    def predict(self, X):
        # Compute predictions
        # Todo
        predictions = (self.decision_function(X) > 0)*1 + (self.decision_function(X) <= 0)*(-1)
        # End todo
        return predictions
```

```python
# Answer
X1 = gaussianMultivar(n, mu=[3, 3], sigma=[8,1], theta=-np.pi/6)
X2 = gaussianMultivar(n, mu=[5, 8], sigma=[8,1], theta=np.pi/3)

X = np.r_[X1, X2]
Y = np.r_[np.ones(X1.shape[0]), -np.ones(X2.shape[0])]

sklda.fit(X, Y)

fda = FisherDA()
fda.fit(X, Y)

plot_frontiere([sklda, fda], X, Y, label=["LDA", "FDA"])

```
